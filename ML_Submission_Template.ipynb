## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.:

Based on the chart experiments, we can formulate three hypothetical statements from the dataset:

Hypothesis 1: There is a significant difference in Annual Premium between individuals who are previously insured and those who are not.

Hypothesis 2: There is a positive correlation between Age and Vintage.

Hypothesis 3: The response rate varies significantly between genders.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.:
Hypothesis Testing for Hypothesis 1:

Null Hypothesis (H0): There is no significant difference in Annual Premium between individuals who are previously insured and those who are not.

Alternative Hypothesis (H1): There is a significant difference in Annual Premium between individuals who are previously insured and those who are not.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_ind

# Split data into previously insured and not previously insured groups
previously_insured = df[df['Previously_Insured'] == 1]['Annual_Premium']
not_previously_insured = df[df['Previously_Insured'] == 0]['Annual_Premium']

# Perform independent t-test
t_stat, p_value = ttest_ind(previously_insured, not_previously_insured)
print("T-statistic:", t_stat)
print("P-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

Answer Here.:

To obtain the p-value, I performed an independent two-sample t-*test* for comparing the means of two independent groups. This test compares the means of the previously insured group and the not previously insured group in terms of Annual Premium. The p-value was calculated based on the t-statistic obtained from this t-test.

##### Why did you choose the specific statistical test?

Answer Here.:

I chose the independent two-sample t-test because it is appropriate for comparing the means of two independent groups when the variable of interest (Annual Premium in this case) is continuous and approximately normally distributed within each group.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.:

Hypothesis 2: There is a positive correlation between Age and Vintage.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Perform Pearson correlation test
from scipy.stats import pearsonr
corr_coefficient, p_value = pearsonr(df['Age'], df['Vintage'])

print("Correlation coefficient:", corr_coefficient)
print("P-value:", p_value)
````


The correlation coefficient obtained is approximately -0.00126, indicating a very weak negative correlation between Age and Vintage. This suggests that there is almost no linear relationship between these two variables.

The p-value obtained is approximately 0.435, which is greater than the commonly used significance level of 0.05. This indicates that there is insufficient evidence to reject the null hypothesis, which states that there is no correlation between Age and Vintage. Therefore, we fail to reject the null hypothesis

##### Which statistical test have you done to obtain P-Value?

Answer Here.:

To obtain the p-value for Hypothesis 2 ("There is a positive correlation between Age and Vintage"), I performed a Pearson correlation test. This test evaluates the strength and direction of the linear relationship between two continuous variables (Age and Vintage). The p-value was calculated based on the correlation coefficient obtained from this test.

##### Why did you choose the specific statistical test?


Answer Here:
Pearson correlation test because it is commonly used to assess the strength and direction of the linear relationship between two continuous variables. In this case, we wanted to determine if there is a positive correlation between Age and Vintage. The Pearson correlation test is suitable for this purpose as it provides both a correlation coefficient and a p-value, allowing us to evaluate the significance of the observed correlation.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.:

Hypothesis 3: The response rate varies significantly between genders.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import chi2_contingency

# Create contingency table for Gender and Response
contingency_table = pd.crosstab(df['Gender'], df['Response'])

# Perform chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)
print("Chi-square statistic:", chi2_stat)
print("P-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

Answer Here.:

To obtain the p-value for Hypothesis 3 ("The response rate varies significantly between genders"), I performed a chi-square test of independence. This test assesses whether there is a significant association between two categorical variables (Gender and Response) in a contingency table format. The p-value was calculated based on the chi-square statistic obtained from this test.

##### Why did you choose the specific statistical test?

Answer Here.:
The chi-square statistic of approximately 1047.71 suggests a strong association between Gender and Response. The extremely small p-value (7.67e-230) indicates that this association is highly unlikely to have occurred by chance. Thus, we reject the null hypothesis and conclude that there is a statistically significant difference in response rates between genders.
